```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
daily_dates = pd.date_range(start='2024-01-01', end='2024-09-01', freq='D')
```

```python
daily_dates = pd.date_range(start='2024-01-01', periods=200, freq='D')
```

```python
daily_dates
```

```python
our_random_values = np.random.uniform(high=40, low=10, size=len(daily_dates))
```

```python
my_random_timeseries = pd.Series(our_random_values, index=daily_dates)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(my_random_timeseries)
plt.show()
```

```python
np.arange(200) * 1
```

```python
trend_data = np.arange(200) * 2
timeseries_trend = pd.Series(trend_data, index=daily_dates)
```

```python
trend_data
```

```python
daily_dates
```

```python
timeseries_trend['2024-01-15':]
```

```python
seasonal_data = 2 * np.sin(2 * np.pi * np.arange(200) / 20)

seasonality = pd.Series(seasonal_data, index=daily_dates)
```

```python
seasonal_data
```

```python
plt.figure(figsize=(16, 8))
plt.plot(timeseries_trend, color='green', alpha=1, label='trend')
plt.plot(my_random_timeseries, color='orange', alpha=1, label='residuals')
plt.plot(seasonality, color='red', label='seasonality')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(13, 4))
# plt.plot(timeseries_trend, color='green', alpha=1, label='trend')
plt.plot(seasonality, color='green', alpha=1, label='seasonality')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
# plt.plot(timeseries_trend, color='green', alpha=1, label='trend')
plt.plot(timeseries_trend+my_random_timeseries+seasonality, color='green', alpha=1, label='combined')
plt.plot(timeseries_trend+my_random_timeseries, color='red', alpha=1, label='combined trend+noise')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
# plt.plot(timeseries_trend, color='green', alpha=1, label='trend')
plt.plot(timeseries_trend*my_random_timeseries*seasonality, color='green', alpha=0.5, label='combined')
plt.plot(timeseries_trend*my_random_timeseries, color='red', alpha=1, label='combined trend+noise')
plt.legend()
plt.show()
```
